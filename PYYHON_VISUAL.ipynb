###Visualizacion de Datos con Python Matplotlib, Seaborn y Ploty

```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/PYTHON VISUAL/inmigrantes_canada.csv')
df.head()
```

```python
df.info()
```

```python
df.set_index('Pais', inplace=True)
df
```

```python
anos = list(map(str, range(1980, 2014)))
anos
```

```python
colombia = df.loc['Colombia',anos]
colombia
```

###Modificacion de tamaño de letras

```python
col_dict = {'Año':colombia.index.tolist(),
            'Inmigrantes':colombia.values.tolist()}
col_dict
```

```python
datos_col = pd.DataFrame(col_dict)
datos_col.tail()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,4))
plt.plot(datos_col['Año'], datos_col['Inmigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010','2013'])
plt.title('Migracion de colombianos a Canada')
plt.xlabel('Años')
plt.ylabel('Numero de Inmigrantes')
plt.show()
```

###Grafico Comparativo de Inmigrantes de Argentina y Brasil

```python
argentina = df.loc['Argentina', anos]
brasil = df.loc['Brasil', anos]
```

```python
col_dict_arg = {'Año':argentina.index.tolist(),
            'Inmigrantes':argentina.values.tolist()}

col_dict_bra = {'Año':brasil.index.tolist(),
            'Inmigrantes':brasil.values.tolist()}
```

```python
datos_arg = pd.DataFrame(col_dict_arg)
datos_arg.tail()
```

```python
datos_bra = pd.DataFrame(col_dict_bra)
datos_bra.tail()
```

```python
x1 = datos_arg['Año']
y1 = datos_arg['Inmigrantes']
x2 = datos_bra['Año']
y2 = datos_bra['Inmigrantes']
```

```python
#plt.figure(figsize=(10,5))
fig, ax = plt.subplots()
ax.plot(x1, y1, label='Argentina')
ax.plot(x2, y2, label='Brasil')
ax.plot(x1, y1, 'o')
ax.plot(x2, y2, 'o')
ax.legend()
plt.xticks(['1980','1985','1990','1995','2000','2005','2010','2013'])
plt.title('Migracion de Argentina y Brasil a Canada')
plt.xlabel('Años')
plt.ylabel('Numero de Inmigrantes')
plt.show()
```

### Otra solucion de grafico comparativo

```python
df_comparacion = df.loc[['Brasil', 'Argentina'], anos]
```

```python
df_comparacion = df_comparacion.T
```

```python
df_comparacion.head()
```

###Grafico de Cajas

```python
#(1,2) significa una fila con dos espacios para graficas
fig, axs=plt.subplots (1,2, figsize=(10,4))

axs[0].plot(datos_col['Año'], datos_col['Inmigrantes'])
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].set_title('Migracion de colombianos hacia Canada\n1980 - 2013')
axs[0].set_xlabel('Años')
axs[0].set_ylabel('Numero de Inmigrantes')
axs[0].grid()

axs[1].boxplot(datos_col['Inmigrantes'])
axs[1].set_title('Boxplot de la inmigracion de colombianos \n hacia Canada 1980 - 2013')
axs[1].set_xlabel('Colombia')
axs[1].set_ylabel('Numero de Inmigrantes')
axs[1].grid()

plt.show()
```

```python
fig, axs=plt.subplots (2,2, figsize=(10,6))
fig.subplots_adjust(hspace=0.5,wspace=0.3)
fig.suptitle('Inmigracion de los 4 mayores paises sudamericanos \n hacia Canada 1980 - 2013')

axs[0,0].plot(df.loc['Colombia', anos])
axs[0,0].set_title('Colombia')

axs[0,1].plot(df.loc['Brasil', anos])
axs[0,1].set_title('Brasil')

axs[1,0].plot(df.loc['Argentina', anos])
axs[1,0].set_title('Argentina')

axs[1,1].plot(df.loc['Perú', anos])
axs[1,1].set_title('Peru')

ymin = 0
ymax = 7000

for ax in axs.ravel():
  ax.set(ylim=(ymin,ymax))

for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.set(xlabel='Años', ylabel='Numero de Inmigrantes')
  ax.grid()


plt.show()


```

###Modificacion al tamaño de Letras

```python
fig,ax = plt.subplots(figsize=(8,3))
plt.style.use('fivethirtyeight')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title('Migracion de colombianos hacia Canada\n1980 - 2013', fontsize=18, loc='left')
ax.set_xlabel('Años',fontsize=14)
ax.set_ylabel('Numero de Inmigrantes',fontsize=14)
ax.plot(datos_col['Año'], datos_col['Inmigrantes'],lw=2,color='g')#marker='o'
ax.yaxis.set_tick_params(labelsize=12)
ax.xaxis.set_tick_params(labelsize=12)
plt.grid(linestyle='--')
plt.show()
```
